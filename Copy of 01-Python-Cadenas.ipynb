<h2 id="strings">Qu es una cadena?</h2>

A continuacin, se muestra una cadena entre comillas dobles.

```python
# Usa las comillas dobles para definir una cadena

"Michael Jackson"
```

Tambin podemos usar comillas simples:

```python
# Usa comillas simples para definir una cadena

'Michael Jackson'
```

Una cadena puede ser una combinacin de espacios y dgitos:

```python
# Dgitos y espacios en una cadena

'1 2 3 4 5 6 '
```

Una cadena tambin puede estar formada por caracteres especiales :

```python
# Caracteres especiales en una cadena

'@#2_#]&*^%$'
```

Podemos imprimir en pantallas una cadena usando la sentencia print:

```python
# Imprimir la cadea

print("hello!")
```

Podemos asignar una cadena como valor de una variable:

```python
# Asignar una cadena a una variable

Name = "Michael Jackson"
Name
```

<h3 id="concat">Concatenar Cadenas</h3>

Podemos concatenar o combinar cadenas usando el smbolo de suma, el resultado ser una nueva cadena que surge de la combinacin de ambas:

```python
# Concatenar dos cadenas

Statement = Name + "is the best"
Statement
```

Para replicar los valores en una cadena simplemente podemos multiplicarla por el numero de veces que deseemos. Para este caso, el numero de veces es tres. El resultado es una nueva cadena, la cual consiste en tres copias de la original:

```python
# Imprime la cadena 3 veces

3 * "Michael Jackson"
```

Se puede crear una nueva cadena aadiendo otra a la original contenida en la variable. Concatenada con una nueva cadena, el resultado es otra diferente, Michael Jackson pas a ser Michael Jackson is the best.

```python
# Concatenar cadenas

Name = "Michael Jackson"
Name = Name + " is the best"
Name
```

<h2 id="escape">Secuencias de Escape</h2>

Una barra invertida representa el inicio de una secuencia de escape. Estas representan cadenas que pudieran ser difciles de introducir. Por ejemplo, una barra invertida seguida de una "n" representa un salto de lnea. En la salida de pantalla tendremos un salto de lnea cada vez que una barra invertida y la letra "n" sea encontrada.

```python
# Secuencia de escape para un Salto de lnea

print(" Michael Jackson \n is the best" )
```

De igual forma, barra invertida "t" aplicar un espacio de tabulacin:

```python
# Secuencia de escape para un espacio de tabulacin

print(" Michael Jackson \t is the best" )
```

En caso de quieras que una barra invertida forme parte de una cadena, utiliza una doble barra invertida:

```python
# Incluir una barra invertida en una cadena

print(" Michael Jackson \\ is the best" )
```

Otra forma de hacer que se muestre la barra invertida de una cadena es colocando antes una r:

```python
# La letra r le dir a python que muestre la cadena tal cual e ignore lo que haya despues de la barra invertida

print(r" Michael Jackson \ is the best" )
```

<h2 id="operations">Operaciones con Cadenas</h2>

En Python existen muchos mtodos para hacer operaciones con cadenas que pueden ser usados para manipular datos. Usaremos algunas operaciones bsicas con cadenas.

Usemos ahora el mtodo <code>upper</code>; este mtodo convierte los caracteres de minsculas a maysculas:

```python
# Convertir todos los caracteres de la cadena en maysculas

A = "Thriller is the sixth studio album"
print("before upper:", A)
B = A.upper()
print("After upper:", B)
```

El mtodo <code>replace</code> reemplaza un segmento de la cadena, p.ej. una sub-cadena por una nueva. Introducimos primero la parte de la cadena que queremos cambiar. En el segundo argumento pondremos el segmento que reemplazara al anterior, el resultado ser una nueva cadena con una de sus partes modificada:

```python
# Reemplazar la vieja sub-cadena por una nueva

A = "Michael Jackson is the best"
B = A.replace('Michael', 'Janet')
B
```

La funcin del mtodo <code>find</code> es encontrar una sub-cadena. El argumento del mtodo es la sub-cadena que deseas encontrar, y el resultado ser el numero del ndice perteneciente al primer carcter de la sub-cadena. Podemos encontrar la sub-cadena <code>jack</code> o <code>el</code>.

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%201/Images/StringsFind.png" width="600" align="center" />

```python
# Encontrar la sub-cadena. El resultado ser unicamente el ndice del primer elemente de la sub-cadena dentro de la cadena principal.

Name = "Michael Jackson"
Name.find('el')
```

```python
# Encotrar la sub-cadena dentro de la cadena.

Name.find('Jack')
```

Si la sub-cadena no se encuentra en la cadena principal el resultado ser -1. Por ejemplo, la cadena 'Jasdfasdasdf' no es una sub-cadena:

```python
# Si no se encuentra la sub-cadena en la cadena principal

Name.find('Jasdfasdasdf')
```
